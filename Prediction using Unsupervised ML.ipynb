```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris_data=pd.read_csv('/home/sai/Downloads/Iris.csv')
iris_data=iris_data.iloc[:,1:]
iris_data.head()
```

## We wish to classify the data into different species based on the first 4 columns. We will not use the species label for training our data. Hence, it is called Unsupervised Machine Learning

### We use 'Elbow Method' to determine optimum number of clusters

```python
from sklearn.cluster import KMeans
X=iris_data.iloc[:,:-1].values
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('Estimating Optimum Number of clusters using Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

As we can see in the above plot, the optimum number of clusters is 3. When we increase the number of clusters beyond this, there is no significant decrease in the Within Cluster Sum of Squares(WCSS).

```python
#Choosing number of clusters as 3
kmeans=KMeans(n_clusters=3,random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

'y_kmeans' gives us the output for the classification based on input data only

### Visualizing the clusters on basis of every pair of features

```python
# Visualising the clusters - On the first two columns
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
